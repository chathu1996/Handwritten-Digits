```python
import tkinter as tk
from PIL import ImageTk,Image,ImageDraw
import numpy as np
import cv2
import joblib

win=tk.Tk()

w,h=500,500
fontButton='Helvetica 20 bold'
fontLabel='Helvetica 24 bold'
count=0

model=joblib.load('sinhala-character-knn.sav')

def eventFunction(event):
    
    x=event.x
    y=event.y
    
    x1=x-30
    y1=y-30
    x2=x+30
    y2=y+30
    
    canvas.create_oval((x1,y1,x2,y2),fill='black')
    imgDraw.ellipse((x1,y1,x2,y2), fill='white')

def save():
    
    global count
    
    imgArray=np.array(img)
    imgArray=cv2.resize(imgArray, (8,8))
    
    cv2.imwrite('data/'+str(count)+'.jpg',imgArray)
    count=count+1
    
def clear():
    
    global img,imgDraw
    
    canvas.delete('all')
    img=Image.new('RGB', (w,h), (0,0,0))
    imgDraw=ImageDraw.Draw(img)
    
def predict():
    
    imgArray=np.array(img)
    imgArray=cv2.cvtColor(imgArray, cv2.COLOR_BGR2GRAY)
    imgArray=cv2.resize(imgArray, (8,8))
    imgArray=np.reshape(imgArray, (1,64))
    imgArray=(imgArray/255.0)*15
    result=model.predict(imgArray)
    labelStatus.config(text="Predicted Digit: "+str(result))
    

canvas=tk.Canvas(win,width=w,height=h,bg='white')
canvas.grid(row=0, column=0, columnspan=4)

buttounSave=tk.Button(win,text="SAVE",bg='light sea green',fg="white",font=fontButton, command=save)
buttounSave.grid(row=1,column=0)

buttounPredict=tk.Button(win,text="PREDICT",bg='blue',fg="white",font=fontButton, command=predict)
buttounPredict.grid(row=1,column=1)

buttounClear=tk.Button(win,text="CLEAR",bg='gold',fg="white",font=fontButton, command=clear)
buttounClear.grid(row=1,column=2)

buttounExit=tk.Button(win,text="EXIT",bg='red',fg="white",font=fontButton, command=win.destroy)
buttounExit.grid(row=1,column=3)

labelStatus=tk.Label(win,text="Predicted Digit: None", bg='white', fg='black', font=fontLabel)
labelStatus.grid(row=2, column=0, columnspan=4)

canvas.bind('<B1-Motion>', eventFunction)
img=Image.new('RGB',(w,h),(0,0,0))
imgDraw=ImageDraw.Draw(img)
win.mainloop()
```
